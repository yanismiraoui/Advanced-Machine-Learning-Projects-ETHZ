```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import normalize, StandardScaler
```

```python
X = pd.read_csv('X_train_processed.csv')
y = pd.read_csv('y_train.csv')
```

```python
X
```

```python
y
```

```python
drop_columns = [
    'HRV_MeanNN',
    'HRV_MedianNN',
    'HRV_LF',
    'HRV_LFHF',
    'HRV_LFn',
    'HRV_SDANN1',
    'HRV_SDANN2',
    'HRV_SDANN5',
    'HRV_SDNNI1',
    'HRV_SDNNI2',
    'HRV_SDNNI5',
    'HRV_ULF',
    'HRV_VLF',   
]
```

```python
X = X.drop(columns=drop_columns)
```

```python
X = X.set_index('id')
y = y.set_index('id')
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X
```

```python
# split the dataset into the training set and test set
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
clf = CatBoostClassifier(learning_rate=0.03, eval_metric='AUC', iterations=1000, task_type="GPU",devices='0:1')
```

```python
clf.fit(X_train, y_train)
```

```python
y_val_pred = clf.predict(X_val)
y_val_pred
```

```python
# view accuracy
from sklearn.metrics import f1_score
accuracy = f1_score(y_val_pred, y_val, average='micro')
print('XGBoost Model accuracy score: {0:0.4f}'.format(accuracy))
```
